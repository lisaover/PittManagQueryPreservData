JOIN the orders DataFrame with the products DataFrame.

```python
orders.loc[:, ['OrderID', 'EmployeeID', 'ProductID', 'Quantity']].copy().\
merge(products.loc[:, ['ProductID', 'CategoryID']].copy(), on='ProductID', how='left')
```

```python
orders.loc[:, ['OrderID', 'EmployeeID', 'ProductID', 'Quantity']].copy().\
merge(products.loc[:, ['ProductID', 'CategoryID']].copy(), on='ProductID', how='left').\
groupby(['EmployeeID']).\
aggregate(num_rows = ('Quantity', 'size'),
          num_categories = ('CategoryID', 'nunique')).\
reset_index()
```

```python
orders.loc[:, ['OrderID', 'EmployeeID', 'ProductID', 'Quantity']].copy().\
merge(products.loc[:, ['ProductID', 'CategoryID']].copy(), on='ProductID', how='left').\
groupby(['EmployeeID']).\
aggregate(num_rows = ('Quantity', 'size'),
          num_categories = ('CategoryID', 'nunique')).\
reset_index().\
merge(employees.loc[:, ['EmployeeID', 'Title']].copy(), on='EmployeeID', how='left')
```

Number of unique Products sold per employee.

```python
orders.loc[:, ['OrderID', 'EmployeeID', 'ProductID', 'Quantity']].copy().\
merge(products.loc[:, ['ProductID', 'CategoryID']].copy(), on='ProductID', how='left').\
groupby(['EmployeeID']).\
aggregate(num_rows = ('Quantity', 'size'),
          num_categories = ('CategoryID', 'nunique'),
          num_products = ('ProductID', 'nunique')).\
reset_index().\
merge(employees.loc[:, ['EmployeeID', 'Title']].copy(), on='EmployeeID', how='left')
```
