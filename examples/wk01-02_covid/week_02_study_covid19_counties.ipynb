Create date related columns for the big data set.

```python
usa['date_dt'] = pd.to_datetime( usa.date )

usa['year'] = usa.date_dt.dt.year

usa['month'] = usa.date_dt.dt.month

usa['day'] = usa.date_dt.dt.day

usa['wday'] = usa.date_dt.dt.day_of_week

usa['quarter'] = usa.date_dt.dt.quarter
```

```python
usa.nunique()
```

## Focus on PA

```python
usa2020.state.value_counts()
```

Use the `.str.contains()` method to find all rows where `state` contains `'Penn'`.

## Counties in PA

Select all rows associated with 3 counties, Allegheny, Centre, and Philadelphia.

```python
pa3 = pa2020.loc[ pa2020.county.isin(['Allegheny', 'Centre', 'Philadelphia']) ].copy()
```

```python
pa3.nunique()
```

Visualizing the data set explains the structure.

```python
sns.relplot( data = pa3, x='date_dt', y='cases', hue='county', kind='line', 
           estimator=None, aspect=2.5, height=5)

plt.show()
```

## Counties in a State

```python
usa2020.groupby(['county']).\
aggregate(num_states = ('state', 'nunique')).\
reset_index()
```

```python
county_only = usa2020.groupby(['county']).\
aggregate(num_states = ('state', 'nunique')).\
reset_index()
```

```python
county_only.loc[ county_only.num_states > 1 ]
```

```python
county_only.loc[ county_only.num_states > 1].sort_values(by='num_states', ascending=False)
```

## Practice Pivoting

```python
pa3.reset_index().loc[:, ['date_dt', 'year', 'month', 'day', 'county', 'cases']].\
pivot(index=['date_dt', 'year', 'month', 'day'], columns='county', values='cases').\
reset_index()
```

Pivot the entire PA data set.

```python
pa2020.reset_index().loc[:, ['date_dt', 'year', 'month', 'day', 'county', 'cases']].\
pivot(index=['date_dt', 'year', 'month', 'day'], columns='county', values='cases').\
reset_index()
```

## Create unique state information

```python
states_2020 = usa2020.groupby(['state']).\
aggregate(num_counties = ('county', 'nunique'),
         num_fips = ('fips', 'nunique'),
         first_date = ('date_dt', 'min'),
         last_date = ('date_dt', 'max')).\
reset_index()
```
