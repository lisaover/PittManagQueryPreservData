#### Stack columns of each price dataset

The four stock prices datasets - open_df, close_df, low_df, and high_df - each have headers that are values not variables. The first set of characters before the "_" are acronyms or symbols that indicate the company, sector, and exchange per the info_df dataset. The second set of characters - open, close, low, and high - describe the price as being the opening, closing, low, or high price, respectively, for a particular stock company, sector, and exchange on a particular day. 

Create list of variable names and then use melt to stack the columns of each stock prices dataset so that the values in the header become part of the observational unit of each row.

## Problem 04: Group and Aggregate the Stock Market Data

#### How many unique days are associated with each MONTH in the data set?

Convert date to datetime and extract the month, day, and year into three new columns. Using that dataset, group by month and year (although the months alone are unique) and count the number of unique days.

```python
stock_prices.info()
```

```python
stock_copy = stock_prices.copy()
stock_copy.head()
```

```python
stock_copy['date_dt'] = pd.to_datetime( stock_copy.date )
stock_copy['year'] = stock_copy.date_dt.dt.year
stock_copy['month'] = stock_copy.date_dt.dt.month
stock_copy['day'] = stock_copy.date_dt.dt.day
stock_copy.drop(['date'], axis=1, inplace=True)
stock_copy.head()
```

```python
stock_copy.groupby(['month', 'year']).\
aggregate(unique_days = ('day', 'nunique')).\
reset_index()
```

#### Calculate the difference between the daily close and daily open prices for each stock.

Using the dataset with datetime values, pivot the table on price_type so daily open and close values are column variables and create two new columns for the difference and the absolute value difference between daily close-to-open prices.

```python
stock_open_close = stock_copy.loc[ stock_copy.price_type.isin(['open','close']), : ].copy()
stock_open_close.head()
```

```python
stock_pivot = stock_open_close.pivot(index=['date_dt', 'year', 'month', 'day', 'company', 'sector', 'exchange'], columns='price_type', values='value').\
reset_index()
stock_pivot.head()
```

```python
stock_pivot['difference'] = stock_pivot['close'] - stock_pivot['open']
stock_pivot.head()
```

```python
stock_pivot['abs_difference'] = abs(stock_pivot['difference'])
stock_pivot['diff_direction'] = np.where(stock_pivot['difference'] >= 0,'gain','loss')
stock_pivot.head()
```

#### What is the maximum daily close-to-open difference for each company?

Using the pivoted dataset with the calculated differences, group by company and obtain the maximum close-to-open difference per company. The table below shows the maximum positive difference, the maximum negative difference, and the maximum absolute value difference with a column that indicates the direction of the absolute value difference as gain or loss.

```python
max_difference_co = stock_pivot.groupby(['company']).\
aggregate(max_positive_diff = ('difference', 'max'),
          max_negative_diff = ('difference', 'min'),
          max_absolute_diff = ('abs_difference', 'max')).\
reset_index()
```

```python
max_difference_co['max_positive_diff'] = max_difference_co['max_positive_diff'].apply(lambda x: "${:.2f}".format((x)))
max_difference_co['max_negative_diff'] = max_difference_co['max_negative_diff'].apply(lambda x: "${:.2f}".format((x)))
max_difference_co['max_absolute_diff'] = max_difference_co['max_absolute_diff'].apply(lambda x: "${:.2f}".format((x)))
max_difference_co['max_abs_direction'] = np.where(max_difference_co['max_absolute_diff'] == max_difference_co['max_positive_diff'],'gain','loss')
max_difference_co
```

#### What is the maximum daily close-to-open difference for each company in each MONTH?

Using the pivoted dataset with the calculated differences, group by company, month, and year and obtain the maximum close-to-open stock price per company per month. The table below shows the maximum positive difference, the maximum negative difference, and the maximum absolute value difference with a column that indicates the direction of the absolute value difference as gain or loss.

```python
max_difference_co_mo = stock_pivot.groupby(['company', 'month', 'year']).\
aggregate(max_positive_diff = ('difference', 'max'),
          max_negative_diff = ('difference', 'min'),
          max_absolute_diff = ('abs_difference', 'max')).\
reset_index()
```

```python
max_difference_co_mo['max_positive_diff'] = max_difference_co_mo['max_positive_diff'].apply(lambda x: "${:.2f}".format((x)))
max_difference_co_mo['max_negative_diff'] = max_difference_co_mo['max_negative_diff'].apply(lambda x: "${:.2f}".format((x)))
max_difference_co_mo['max_absolute_diff'] = max_difference_co_mo['max_absolute_diff'].apply(lambda x: "${:.2f}".format((x)))
max_difference_co_mo['max_abs_direction'] = np.where(max_difference_co_mo['max_absolute_diff'] == max_difference_co_mo['max_positive_diff'],'gain','loss')
max_difference_co_mo
```

#### What is the average daily close-to-open difference for each company in each MONTH?

Using the pivoted dataset with the calculated differences, group by company, month, and year and obtain the average close-to-open difference per company per month.

```python
avg_diff_co_mo = stock_pivot.groupby(['company', 'month', 'year']).\
aggregate(mean_difference = ('difference', 'mean')).\
reset_index()
```

```python
avg_diff_co_mo['mean_difference'] = avg_diff_co_mo['mean_difference'].apply(lambda x: "${:.2f}".format((x)))
avg_diff_co_mo
```
